this section we are merging some columns into one and selection only the necessary columns to be used while merging to the main occupation dataset

```python
#merging data value and element for interest
interests = interests.rename(columns={'Element Name': 'RIASEC','Data Value': 'RIASEC_Score'})
interests['RIASEC_Value'] = interests['RIASEC'] + " (" + interests['RIASEC_Score'].astype(str) + ")"
career_interests= interests.groupby('O*NET-SOC Code')['RIASEC_Value'].apply(lambda x: sorted(set(x))).reset_index()
career_interests.columns = ['O*NET-SOC Code', 'RIASEC']

#merging element and data value for abilities
abilities['Ability_Value'] = abilities['Element Name'] + " (" + abilities['Data Value'].astype(str) + ")"
career_abilities = abilities.groupby('O*NET-SOC Code')['Ability_Value'].apply(lambda x: sorted(set(x))).reset_index()
career_abilities.columns = ['O*NET-SOC Code', 'Abilities']

#skills grouping
skills['Skill_Value'] = skills['Element Name'] + " (" + skills['Data Value'].astype(str) + ")"
career_skills = skills.groupby('O*NET-SOC Code')['Skill_Value'].apply(lambda x: sorted(set(x))).reset_index()
career_skills.columns = ['O*NET-SOC Code', 'Skills']

#emerging task grouping
emerging_task = emerging[['O*NET-SOC Code', 'Category']].dropna()
emerging_task = emerging_task.rename(columns={'Category': 'Education_Level'})
emerging_careers = emerging_task.groupby('O*NET-SOC Code')['Education_Level'].apply(lambda x: sorted(set(x))).reset_index()
emerging_careers.columns = ['O*NET-SOC Code', 'Emerging_Tasks']
```

These section we want to create new columns on the education datasets. Element column has differnt row sections which can be converted to columns and assigned categories after mapping them out

```python
#filtering only relevant levels
Career_Levels = education[education['Element Name'].isin([
    'Required Level of Education',
    'On-the-Job Training',
    'Related Work Experience',
    'On-Site or In-Plant Training',
    'Job-related Apprenticeship'
])]
```

```python
education_level_category = {
    1: "Less than High School",
    2: "High School Diploma or equivalent",
    3: "Post-Secondary Certificate",
    4: "Some College Courses",
    5: "Associate's Degree",
    6: "Bachelor's Degree",
    7: "Post-Baccalaureate Certificate",
    8: "Master's Degree",
    9: "Post-Master's Certificate",
    10: "First Professional Degree",
    11: "Doctoral Degree",
    12: "Post-Doctoral Training"
}

ojt_category = {
    1: "None or short demonstration",
    2: "Up to 1 month",
    3: "Over 1 to 3 months",
    4: "Over 3 to 6 months",
    5: "Over 6 months to 1 year",
    6: "Over 1 to 2 years",
    7: "Over 2 to 4 years",
    8: "Over 4 to 10 years",
    9: "Over 10 years"
}

work_experience_category = {
    1: "None",
    2: "Up to 1 month",
    3: "Over 1 to 3 months",
    4: "Over 3 to 6 months",
    5: "Over 6 months to 1 year",
    6: "Over 1 to 2 years",
    7: "Over 2 to 4 years",
    8: "Over 4 to 6 years",
    9: "Over 6 to 8 years",
    10: "Over 8 to 10 years",
    11: "Over 10 years"
}

onsite_category = {
    1: "None",
    2: "Up to 1 month",
    3: "Over 1 to 3 months",
    4: "Over 3 to 6 months",
    5: "Over 6 months to 1 year",
    6: "Over 1 to 2 years",
    7: "Over 2 to 4 years",
    8: "Over 4 to 10 years",
    9: "Over 10 years"
}

apprenticeship_importance_category = {
    1: "Less Important",
    2: "Somewhat Important",
    3: "Moderately Important",
    4: "Very Important",
    5: "Most Important"
}
```

This section we are going to choose the maximum level and value reguired per career

```python
#pivot_table() cats
career_level_category = Career_Levels.pivot_table(
    index='O*NET-SOC Code',
    columns='Element Name',
    values='Category',
    aggfunc='max'
).reset_index()

career_level_category.columns.name = None
#pivot_table() values
career_level_value = Career_Levels.pivot_table(
    index='O*NET-SOC Code',
    columns='Element Name',
    values='Data Value',
    aggfunc='max'
).reset_index()


career_level_value.columns.name = None
```

```python
#renaming our columns
career_level_value = career_level_value.rename(columns={
    'Required Level of Education': 'Education_Level_Value',
    'On-the-Job Training': 'OJT_Value',
    'Related Work Experience': 'Work_Exp_Value',
    'On-Site or In-Plant Training': 'InPlant_Value',
    'Job-related Apprenticeship': 'Internship_Value'
})

career_level_category = career_level_category.rename(columns={
    'Required Level of Education': 'Education_Level',
    'On-the-Job Training': 'OJT',
    'Related Work Experience': 'Work_Experience',
    'On-Site or In-Plant Training': 'InPlant_Training',
    'Job-related Apprenticeship': 'Internship'
})
```

```python
print(career_level_value.columns.tolist())
```

```python
#created new matrix for internship
career_level_value['Internship'] = career_level_value['Internship_Value'].map(apprenticeship_importance_category)

#merging career level category and career level value
career_level_merged = pd.merge(career_level_category, career_level_value, on='O*NET-SOC Code', how='left')

#map decsriptions

career_level_merged['Education_Level'] = career_level_merged['Education_Level'].map(education_level_category)
career_level_merged['OJT'] = career_level_merged['OJT'].map(ojt_category)
career_level_merged['Work_Experience'] = career_level_merged['Work_Experience'].map(work_experience_category)
career_level_merged['InPlant_Training'] = career_level_merged['InPlant_Training'].map(onsite_category)
career_level_merged['Internship'] = career_level_merged['Internship'].map(apprenticeship_importance_category)

```

```python
Education_final = career_level_merged[[
    'O*NET-SOC Code',
    'Education_Level', 'Education_Level_Value',
    'OJT', 'OJT_Value',
    'Work_Experience', 'Work_Exp_Value',
    'InPlant_Training', 'InPlant_Value',
    'Internship', 'Internship_Value'
]]

# Step 9: Preview
print(Education_final.head(10))
```

This section we are going to merge the different datasets into one. we are creating a copy called careers_df

```python
#creating copy
careers_df = occupations.copy()

# merging copy of occupations to interest
careers_df = pd.merge(careers_df, career_interests, on='O*NET-SOC Code', how='left')

#merging career_df+abilities
careers_df = pd.merge(careers_df, career_abilities, on='O*NET-SOC Code', how='left')

#merging career_df+abilities+skills
careers_df = pd.merge(careers_df, career_skills, on='O*NET-SOC Code', how='left')
#merging career_df+abilities+skills+education
careers_df = pd.merge(careers_df, Education_final, on='O*NET-SOC Code', how='left')

careers_df.head(10)
```

```python
#joining occupation and emerging tasks
occupation_merge = pd.concat([careers_df, emerging_careers], ignore_index=True)

occupations_merge.tail()
```
